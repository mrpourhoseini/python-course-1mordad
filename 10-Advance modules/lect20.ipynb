# datetime

```python
from datetime import datetime
```

```python
help(datetime)
```

```python
t = datetime.now()
```

```python
print(t)
print(t.year)
print(t.month)
print(t.day)
print(t.hour)
print(t.minute)
print(t.second)
print(t.microsecond)
print(t.ctime())
```

```python
today = datetime.today()
```

```python
print(today)
print(today.ctime())
```
